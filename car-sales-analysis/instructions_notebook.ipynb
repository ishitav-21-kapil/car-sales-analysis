```python
# Import the libraries needed for working with data and creating plots.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a CSV file into a DataFrame.
car_sales_data = pd.read_csv("cars.csv")
car_sales_data
```

```python
# Show the first few rows to get an idea of how the data looks.
car_sales_data.head()
```

```python
# Look at column names, data types, and overall information.
car_sales_data.columns
car_sales_data.dtypes
car_sales_data.info()
```

```python
# Generate a summary of numerical data.
# Observe values like average, minimum, and maximum.

car_sales_data.describe()
```

```python
car_sales_data[["Price","Mileage"]].mean()
car_sales_data[["Price","Mileage"]].median()
car_sales_data[["Price","Mileage"]].mode()
```

```python
# Look for any unnecessary columns such as unnamed index columns
# and remove them if they are not useful.
car_sales_data.columns
```

```python
# Check if there are any missing values in the dataset.
car_sales_data.isnull().any()
```

```python
# Create a small custom DataFrame using your own data
# such as car names and colors.
rows = pd.Series(["Toyato","Honda","BMW","Mercedes","Tesla"])
column = pd.Series(["Red","Blue","Green","White","Black"])
doors = pd.Series([2,4,3,5,2])
prices = pd.Series([230000,45000,56000,24000,129800])
car_data = pd.DataFrame({"Brands":rows,"Colors":column,"Doors":doors,"Prices($)":prices})
car_data
```

```python
# Perform basic analysis on important columns.
car_data["Prices($)"].median()
car_data["Prices($)"].max()
car_data["Prices($)"].min()
car_data["Prices($)"].mean()
```

```python
car_data.columns
```

```python
# Create a simple plot to understand how values are distributed.
# A histogram of price is a good starting point.
# Add titles and labels to the plot so it is easy to understand.
car_data.plot(x="Brands", y="Prices($)", kind="bar")

plt.title("Car Prices by Brand")
plt.xlabel("Car Brand")
plt.ylabel("Price")

plt.show()
```

```python
# Save the cleaned dataset(car_sales_data) into a new CSV file.
car_sales_data
```

```python
# Save the cleaned dataset into a new CSV file.
car_data.to_csv('car-sales-data.csv', index=False)
```
